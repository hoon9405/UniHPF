```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df = pd.read_csv('single.csv')
```

```python
def sed(x):
    return np.mean(x), np.std(x)/np.sqrt(5)
df = df.groupby('Task').agg(sed)

df = df.reindex(index=['Mort', 'Los3', 'Los7', 'Readm', 'fi_ac', 'im_disch', 'dx'])
df = df.rename({'fi_ac':'Fi_ac', 'im_disch':'Im_disch', 'dx':'Dx', 'Los3':'LOS3', 'Los7':'LOS7'})
df
```

```python
tasks = ['Mort', 'LOS3', 'LOS7', 'Readm', 'Fi_ac', 'Im_disch', 'Dx']
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plt.figure(figsize=(16, 10))
colors = ['black', 'black', 'blue', 'blue', 'blue', 'red']
markers = ['^', 'o', '^', 's', 'D', 'o']
for i, dataset in enumerate(['MIMIC-III', 'eICU', 'MIMIC-IV']):
    for j, task in enumerate(df.index):
        plt.subplot(3, 7, 7*i+j+1)
        data = df.loc[task].values[6*i:6*i+6]
        point, err = zip(*data)
        # Marker/Color
        print(point)
        point = np.array(point)[[0,1,2,4,3,5]]
        err = np.array(err)[[0,1,2,4,3,5]]
        for k in range(6):
            if i==2 and k<2:
                continue
            plt.errorbar(k, point[k], yerr=err[k], fmt=markers[k], color=colors[k])
        if i==2:
            data = data[2:]
        data_min = min([i-j for i,j in data]) //0.02 *0.02
        data_max = max([i+j for i,j in data]) //0.02*0.02 +0.021

        ticks = np.arange(data_min, data_max, 0.02)
        if data_max - data_min > 0.15:
            ticks = np.arange(data_min, data_max, 0.04)
        if data_max - data_min <0.04:
            ticks = np.arange(data_min, data_max, 0.01)
        plt.yticks(ticks)
        plt.xticks(np.arange(6))
        plt.tick_params(
            axis='x',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            bottom=False,      # ticks along the bottom edge are off
            top=False,         # ticks along the top edge are off
            labelbottom=False)
        if i==0:
            plt.title(task)
        if j==0:
            plt.ylabel(dataset)
        plt.tight_layout()
        plt.savefig('vldb_single.pdf',format='pdf')
# Benchmakr, Ours-Benchmark, codeemb, descemb, google, ours
```
